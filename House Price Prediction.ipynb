```python
def remove_outliers_sqft(obj):
    output = pd.DataFrame()
    for key,sub in obj.groupby('Locality'):
        m = np.mean(sub.Per_Sqft)
        s = np.std(sub.Per_Sqft)
        gen_df = sub[(sub.Per_Sqft > (m-s)) & (sub.Per_Sqft <= (m+s))]
        output = pd.concat([output,gen_df],ignore_index=True)
        return output
```

```python
def bhk_outlier(df):
    exclude_index = np.array([])
    for location,location_def in df.groupby('Locality'):
        bhk_stats = {}
        for bhk , bhk_def in location_def.groupby('BHK'):
            bhk_stats[bhk] = {
                'mean' : np.mean(bhk_def.Per_Sqft),
                'std' : np.std(bhk_def.Per_Sqft),
                'count' : bhk_def.shape[0]
            }
        for bhk,bhk_def in location_def.groupby('BHK'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>3:
                exclude_index = np.append(exclude_index,bhk_def[bhk_def.Per_Sqft<(stats['mean'])].index.values)
    return df.drop(exclude_index,axis="index")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train ,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
import pickle
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse=False),['Locality']),remainder="passthrough")
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression(normalize=True)
```

```python
pipe = make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred_lr)
pickle.dump(pipe,open('LinearModel.pkl','wb'))
```

```python
# Lasso Function
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lasso = pipe.predict(x_test)
r2_score(y_test,y_pred_lasso)
```

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(x_train,y_train)
```
