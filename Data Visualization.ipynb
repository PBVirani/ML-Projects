```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
data = pd.read_csv('pokemon.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
# countplot is used for create a bar chart
sb.countplot(data=data,y="type_1")
```

```python
color = sb.color_palette()[0]
#Color Palette function is used for give color in bar chart
sb.countplot(data=data,y="type_1",color=color)
```

```python
# Xtricks function is use to rotate label in Chart
sb.countplot(data=data,x="type_1")
plt.xticks(rotation=90)
```

```python
# order is use for show data in perticular order
type = data['type_1'].value_counts().index
sb.countplot(data=data,x="type_1",order=type)
plt.xticks(rotation=90)
```

```python
# Barplot is use for unorder data show
# In Barplot(x,y) fixed arguments
na_counts = data.isna().sum()
base_color = sb.color_palette()[0]
sb.barplot(na_counts,na_counts.index.values, color = base_color)
```

```python
# Pie function is use for pie chart
# In Pie Chart Pie(values_count,labels,startangle.counterclock) 4 Agruments are mainly.
sorted_counts = data['generation_id'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90,
        counterclock = False);
plt.axis('square');
```

```python
# In Pie Chart wedgeprops is accept dictionary type values and width must be between 0 to 1.
sorted_counts = data['generation_id'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90,
        counterclock = False ,wedgeprops={'width':.3});
plt.axis('square');
```

```python
# Hist function use for create Histogram
plt.hist(data = data, x = 'speed')
```

```python
# Bin is use for divide data in number of bins
plt.hist(data = data, x = 'speed',bins=20)
```

```python
# Here First we Calculat the max speed and after divide all speed value in 5 of bin
bins = np.arange(0, data['speed'].max()+1, 5)
plt.hist(data = data, x = 'speed', bins = bins)
```

```python
plt.figure(figsize = [10, 5]) # larger figure size for subplots

# histogram on left, example of too-large bin size
plt.subplot(1, 2, 1) # 1 row, 2 cols, subplot 1
bin_edges = np.arange(0, data['speed'].max()+1, +5)
plt.hist(data = data, x = 'speed', bins = bin_edges);

# histogram on right, example of too-small bin size
plt.subplot(1, 2, 2) # 1 row, 2 cols, subplot 2
bin_edges = np.arange(0, data['speed'].max()++1, 1)
plt.hist(data = data, x = 'speed', bins = bin_edges);
```

```python
# Displot is used to show data like histogram
sb.displot(data['speed'])
```

```python
# Distplot is use for show data in histogram with max bin line
sb.distplot(data['speed'])
```

```python
# If you give False in kde than line are not display
# hist_kws is use for color
sb.distplot(data['speed'],kde=False,hist_kws = {'alpha' : 1})
```
